Buid a sample dataset containing similar distribution of normal transaction and the fraudulent transaction 

Number of Fraudulent Transactions====> 492
so we are taking 492 normal transaction for balance data

```python
legit=df_Normal.sample(n=492) #it randomly sample the data
```

```python
legit
```

```python
new_df= concat(([legit,df_Fraud]), axis=0)
```

```python
new_df
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
X= new_df.drop(columns='Class',axis=1)
Y= new_df["Class"]
```

```python
X
```

split it into Training Data and Testing Data

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
X_test
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_pre= model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_pre,Y_train)
```

```python
training_data_accuracy
```

```python
X_test_pre= model.predict(X_test)
```

```python
testing_data_accuracy= accuracy_score(X_test_pre, Y_test)
testing_data_accuracy
```

Model prediction is good as it shows 93.4% accuracy
